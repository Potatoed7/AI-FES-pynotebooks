# Plotting with Numpy

If we don't have dataframes, we can also plot the old fashioned way: by specifying our x and y coordinates. Here, we pass our data into the plt.plot() function:

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

x = np.array(range(1,15))
y = x ** 2 + 5  #this can be dependent on x, but could also be an array as well
plt.plot(x,y,"ob")
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.show()
```

You can also get very fancy by customising all the details of the plot. Here we use a dataset on the Titanic passengers:

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/iamamangosteen/aimlnotebooks/main/titanic.csv')
data = data.dropna()

data.head()
```

```python
# convert gender data to numeric data: female = 1, male -1
data['GenderNum'] = data.Sex.apply(lambda name: (name == 'female')*1 - (name == 'male')*1)

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
aggregation = data.groupby(['GenderNum', 'Survived'])['Name'].count().to_dict()

fig, ax = plt.subplots(2, 2, figsize=(8,7))

for i, gender in enumerate([-1, 1]):
    for survived in [0, 1]:

        # Define x, y
        x_pos = survived
        y = aggregation[(gender, survived)]

        # plot bar
        ax[i, 0].bar(x_pos, y, width=0.1, edgecolor='black', color=(survived, 0.8, 0.9), alpha=0.8)

    gender_label = (gender==-1)*'male'.upper() + (gender==1)*'female'.upper()
    ax[i, 0].set_title('Survival Count By Gender Category: ' + gender_label)
    ax[i, 0].legend(['died', 'survived'])
    ax[i, 0].grid(axis='y')
    ax[i, 0].set_ylim(0, 500)

# pie chart
colors = [(0, 0.8, 0.9), (1, 0.8, 0.9)]
for i, gender in enumerate([-1, 1]):

    labels = ['died', 'survived']
    sizes = []
    for survived in [0, 1]:
        sizes.append(aggregation[(gender, survived)])
    # plot pie
    ax[i, 1].pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, colors=colors)

    gender_label = (gender==-1)*'male'.upper() + (gender==1)*'female'.upper()
    ax[i, 1].set_title('Survival Count By Gender Category: ' + gender_label)
    ax[i, 1].legend(['died', 'survived'])
    ax[i, 1].grid(axis='y')

fig.tight_layout()
fig.patch.set_facecolor((0.8, 0.8, 0.6))
plt.show()
```

```python
data['Survival'] = data.Survived.map({0 : 'Died', 1 : 'Survived'})
data['Class'] = data.Pclass.map({1 : '1st Class', 2 : '2nd Class', 3 : '3rd Class'})
data['Gender'] = data.Sex.map({'female' : 'Female', 'male' : 'Male'})
# Replace blanks with NaN
data['Embarked'].replace(r'\s+', np.nan, regex=True).head()
data['Ports'] = data.Embarked.map({'S' : 'Southhampton', 'C' : 'Cherbourg', 'Q' : 'Queenstown', np.nan : 'unknown'})

# Mosaic Chart
plt.rc('figure', figsize=(17, 5))
from statsmodels.graphics.mosaicplot import mosaic
mosaic(data,['Survival', 'Class', 'Gender'], axes_label=False, title='Survival: Red=Died, Green=Survived')
plt.xlabel('Gender: Male & Female')
plt.ylabel('Passenger Class: 1st, 2nd & 3rd Class')
plt.show()
```

```python
# Maintain Age and create Age_Fill (populate missing ages)
data['Age_Fill'] = data['Age']

# Populate Age_Fill
data['Age_Fill'] = data['Age_Fill'] \
    .groupby([data['Pclass'], data['Survived'], data['Sex']], group_keys=False) \
    .apply(lambda x: x.fillna(x.mean()))

# Set figure size
plt.rc('figure', figsize=(10, 10))

# Setup a figue of plots

df1 = data[data['Survived'] == 0]['Age']
df2 = data[data['Survived'] == 1]['Age']
df3 = data[data['Survived'] == 0]['Age_Fill']
df4 = data[data['Survived'] == 1]['Age_Fill']

plt.legend(('Died', 'Survived'), loc='best')
plt.title('Survivors by Age Group')
plt.xlabel('Age')
plt.ylabel('Count')

max_age = max(data['Age_Fill'])

plt.subplot(311)
plt.hist([df1, df2],
             bins=8,
             range=(1, max_age),
             stacked=True)

plt.legend(('Died', 'Survived'), loc='best')
plt.title('Survivors by Age Group (not filled)')
plt.ylabel('Count')

plt.subplot(312)
plt.hist([df3, df4],
             bins=8,
             range=(1, max_age),
             stacked=True)

plt.legend(('Died', 'Survived'), loc='best')
plt.title('Survivors by Age Group (filled)')
plt.xlabel('Age')
plt.ylabel('Count')

plt.show()
```

### <font color="red">Exercise 4: Plotting Multiple Countries</font>

Plot the GDP per capita growth of Singapore alongside two other ASEAN countries from 1952 to 2007 using plotly:

### <font color="red">Plotly Example</font>

So far we've only plotted scatter plots with plotly. What about a bar graph? You'll notice we use 'plotly.graph_objs.Scatter' in our above examples. To plot a bar graph, we'll use **plotly.graph_objs.Bar**. We'll be using the CSV located on plotly's github below:

```python
#Here's some starter code:

df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/school_earnings.csv")
df.head()
#This is a dataset hosted by plotly for practice use.

data = [go.Bar(x=df.School, y=df.Gap)]

fig = go.Figure(data = data)
fig.show()
```

Or perhaps you want to make it a bit more fancy by including multiple bars per school:

```python
import plotly.graph_objs as go

df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/school_earnings.csv")

trace_women = go.Bar(x=df.School,
                  y=df.Women,
                  name='Women',
                  marker=dict(color='#ffcdd2'))

trace_men = go.Bar(x=df.School,
                y=df.Men,
                name='Men',
                marker=dict(color='#A2D5F2'))

trace_gap = go.Bar(x=df.School,
                y=df.Gap,
                name='Gap',
                marker=dict(color='#59606D'))

data = [trace_women, trace_men, trace_gap]

layout = go.Layout(title="Average Earnings for Graduates",
                xaxis=dict(title='School'),
                yaxis=dict(title='Salary (in thousands)'))

fig = go.Figure(data = data, layout = layout)
fig.show()
```

```python
import plotly
import pandas as pd
df = pd.read_csv('http://www.stat.ubc.ca/~jenny/notOcto/STAT545A/examples/gapminder/data/gapminderDataFiveYear.txt', sep='\t')
df2007 = df[df.year==2007]
df1952 = df[df.year==1952]
df.head(2)

fig = {
    'data': [
        {
            'x': df2007.gdpPercap,
            'y': df2007.lifeExp,
            'text': df2007.country,
            'mode': 'markers',
            'name': '2007'},
        {
            'x': df1952.gdpPercap,
            'y': df1952.lifeExp,
            'text': "df1952.country",
            'mode': 'markers',
            'name': '1952'}
    ],
    'layout': {
        'xaxis': {'title': 'GDP per Capita', 'type': 'log'},
        'yaxis': {'title': "Life Expectancy"}
    }
}

go.Figure(fig).show()
```

### Stacked Bar Graphs

```python
import numpy as np

N = 20
x = np.linspace(1, 10, N)
y = np.random.randn(N)+3
y2 = np.random.randn(N)+6
y3 = np.random.randn(N)+9
y4 = np.random.randn(N)+12
df = pd.DataFrame({'x': x, 'y': y, 'y2':y2, 'y3':y3, 'y4':y4})
df.head()

data = [
    go.Bar(
        x=df['x'], # assign x as the dataframe column 'x'
        y=df['y']
    ),
    go.Bar(
        x=df['x'],
        y=df['y2']
    ),
    go.Bar(
        x=df['x'],
        y=df['y3']
    ),
    go.Bar(
        x=df['x'],
        y=df['y4']
    )

]

layout = go.Layout(
    barmode='stack',
    title='Stacked Bar with Pandas'
)

fig = go.Figure(data=data, layout=layout)

fig.show()
```
